## Text Preprocessing(Clean data)

```python
import pandas as pd  
import string
```

```python
df=pd.read_csv(r"C:\Users\Admin\Desktop\text preprocessing Nlp\Twitter sentiments project\Twitter Sentiments (1).csv")
df.head()
```

```python
df=df.drop(columns=['id','label'],axis=1)
df.head()
```

## convert to lowercase

```python
df['clean_text']=df['tweet'].str.lower()
df.head()
```

## remove of punctuations

```python
def remove_punctuations(text):
  punctuations=string.punctuation
  return text.translate(str.maketrans('','',punctuations))

df['clean_text']=df['clean_text'].apply(lambda text: remove_punctuations(text))
df.head()
```

## remove of stopwords

```python
import nltk  
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
def remove_stopwords(text):
  return " ".join([word for word in str(text).split() if word not in stop_words])

df['clean_text']=df['clean_text'].apply(lambda text: remove_stopwords(text))
df
```

## Remove Of Frequent Words

```python
import collections
from collections import Counter
word_counter=Counter()
for text in df['clean_text'].values:
  for word in text.split():
    word_counter[word]+=1
    most_common=word_counter.most_common(10)

word_counter.most_common(10)
```

```python
frequent_words=set(word for (word,count) in word_counter.most_common(3))
def remove_frequentwords(text):
  return " ".join([word for word in str(text).split() if word not in frequent_words])

df['clean_text']=df['clean_text'].apply(lambda text: remove_frequentwords(text))
df
```

## remove of rare words

```python
rare_words=set(word for (word,count) in word_counter.most_common()[:-10:-1])
rare_words
```

```python
def remove_rarewords(text):
  return " ".join([word for word in str(text).split() if word not in rare_words])

df['clean_text']=df['clean_text'].apply(lambda text: remove_rarewords(text))
df
```

## Remove of Special Characters

```python
import re
def remove_specialcharacters(text):
  text=re.sub('[^a-zA-Z0-9\s]','',text)
  text=re.sub('\s+',' ',text)
  return text

df['clean_text']=df['clean_text'].apply(lambda text: remove_specialcharacters(text))
df
```

## Stemming

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
def stem_words(text):
  return " ".join([stemmer.stem(word) for word in text.split()])

df['clean_text']=df['clean_text'].apply(lambda text: stem_words(text))
df
```

## Lematization & Pos tagging

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

def lemmatize_words(text):
    # find pos tags
    pos_text = pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])
df['lemmatized_text'] = df['clean_text'].apply(lambda x: lemmatize_words(x))
df.head()
```

## Remove of URLs

```python
text = "https://www.hackersrealm.net is the URL of the channel Hackers Realm"
import re
def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
remove_urls(text)
```

```python
df['clean_text']=df['clean_text'].apply(remove_urls)
df.head()
```

## Remove Of HTML Tages

```python
text = "<html><body> <h1>Hackers Realm</h1> <p>This is NLP text preprocessing tutorial</p> </body></html>"
import re
def remove_html(text):
    return re.sub(r'<.*?>', '', text)
remove_html(text)
```

```python
df['clean_text']=df['clean_text'].apply(remove_html)
df.head()
```

## Remove Spelling Correction

```python
text = 'natur is a beuty'
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()
def correct_spellings(text):
    corrected_text = []
    misspelled_words = spell.unknown(text.split())
    #print(misspelled_words)
    for word in text.split():
        if word in misspelled_words:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
```
